```python
from bs4 import BeautifulSoup
import requests
import random

class hollywoodStars:
    def top_stars():
        html_stars = requests.get('https://www.imdb.com/list/ls022928819/').text
        star_soup = BeautifulSoup(html_stars, 'html.parser')

        stars_list = []

        stars = star_soup.find_all('div', class_='lister-item mode-detail')
        for star in stars:
            star = star.div.a.img['alt']
            stars_list.append(star)
        
        return stars_list

    def walk_of_fame():
        html_text = requests.get('https://en.wikipedia.org/wiki/List_of_stars_on_the_Hollywood_Walk_of_Fame').text
        fame_soup = BeautifulSoup(html_text, 'html.parser')

        actor_lst = hollywoodStars.top_stars()
        actor_dict = {}
        all_a = fame_soup.find_all('a')
        for a in all_a:
            if a.text in actor_lst:
                html = ('https://en.wikipedia.org' + a['href'])
                actor_dict[a.text] = html

        return actor_dict

    def get_hints():
        hollywood_dict = hollywoodStars.walk_of_fame()
        global pick_star
        pick_star = random.choice(list(hollywood_dict.items()))

        actor_text = requests.get(f"{pick_star[1]}").text
        hint_soup = BeautifulSoup(actor_text, 'html.parser')

        categories = hint_soup.find('div', id='catlinks')
        all_cate = categories.find_all('li')
        global hint_list
        hint_list = []

        for cate in all_cate:
            if len(hint_list) <= 25:
                hint_list.append(cate.text)
            else:
                break

        for i in hint_list:
            if pick_star[0] in i or '21st-century' in i:
                hint_list.remove(i)
        
        return hint_list

    def question_game():
        greeting = "Who are some of the most popular Hollywood Stars on the Walk of Fame! \nGuess an actor and see if we are thinking of the same person."
        user = input("Guess an actor.")

        print(greeting)
        print(user)
        
        lives = 10

        while user != pick_star[0]:
            if user != pick_star[0]:
                print(f'Sorry, that is not who I was thinking. You have {lives} guesses left.')
                hint = random.choice(hint_list)
                print(f'Here is a hint: {hint}')
                hint_list.remove(hint)
                lives = lives - 1
                user = input("Guess again.")
                print(user)

            if user == pick_star[0]:
                print("Good Guess! You know your Hollywood Stars, Winner!")
                break

            if lives == 0:
                print(f"I'm sorry I was thinking of {pick_star[0]}")
                break

        print("Goodbye")
        
hollywoodStars.get_hints()
hollywoodStars.question_game()
```
